```python
import nltk
from nltk.corpus import stopwords
def tokenize_text(text):
    tokens = []
    for sent in nltk.sent_tokenize(text):
        for word in nltk.word_tokenize(sent):
            #if len(word) < 0:
            if len(word) <= 0:
                continue
            tokens.append(word.lower())
    return tokens
train_tagged = train_df_cleaned.apply(
    lambda r: TaggedDocument(words=tokenize_text(r['crimeaditionalinfo']), tags=[r.category]), axis=1)
test_tagged = test_df_cleaned.apply(
    lambda r: TaggedDocument(words=tokenize_text(r['crimeaditionalinfo']), tags=[r.category]), axis=1)

# The maximum number of words to be used. (most frequent)
max_fatures = 500000

# Max number of words in each complaint.
MAX_SEQUENCE_LENGTH = 50

#tokenizer = Tokenizer(num_words=max_fatures, split=' ')
tokenizer = Tokenizer(num_words=max_fatures, split=' ', filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(train_df_cleaned['crimeaditionalinfo'].values)
X = tokenizer.texts_to_sequences(train_df_cleaned['crimeaditionalinfo'].values)
X = pad_sequences(X)
print('Found %s unique tokens.' % len(X))

#tokenizer = Tokenizer(num_words=max_fatures, split=' ')
tokenizer = Tokenizer(num_words=max_fatures, split=' ', filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(test_df_cleaned['crimeaditionalinfo'].values)
Y = tokenizer.texts_to_sequences(test_df_cleaned['crimeaditionalinfo'].values)
Y = pad_sequences(Y)
print('Found %s unique tokens.' % len(Y))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten


# One-hot encoding for the 'category' column in the training set
Y = pd.get_dummies(train_df_cleaned['category']).values

# Prepare text data in 'sub_category' and 'crimeaditionalinfo' columns
# Convert to strings and handle NaN values
texts = (train_df_cleaned['sub_category'].fillna('') + " " + train_df_cleaned['crimeaditionalinfo'].fillna('')).astype(str)

# Tokenize the text data
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(texts)
X = tokenizer.texts_to_sequences(texts)
X = pad_sequences(X, maxlen=100)  # Padding to ensure consistent input shape

# Splitting the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=42)

# Build a simple model as an example
model = Sequential([
    Embedding(input_dim=10000, output_dim=64, input_length=100),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(Y_train.shape[1], activation='softmax')  # Output layer with number of classes
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Fit the model
batch_size = 32
history = model.fit(X_train, Y_train, epochs=50, batch_size=batch_size, verbose=2)

# Display shapes (optional, for verification)
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)
print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)
```

```python
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
plt.savefig('model_accuracy.png')

# summarize history for loss
plt.plot(history.history['loss'])
#plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
plt.savefig('model_loss.png')
```

```python
# evaluate the model
_, train_acc = model.evaluate(X_train, Y_train, verbose=2)
_, test_acc = model.evaluate(X_test, Y_test, verbose=2)
print('Train: %.3f, Test: %.4f' % (train_acc, test_acc))
```
